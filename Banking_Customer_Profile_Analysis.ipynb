# Banking Customer Profile Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df= pd.read_csv("Banking.csv")
```

```python
df['GenderId'].unique()
```

```python
df['Risk Weighting'].unique()
```

```python
df1= df.groupby('Fee Structure')['Client ID'].count().reset_index()
df1.columns= ['Fee Structure', 'Num of Clients']
df1.head()
```

```python
df2= df.groupby('Loyalty Classification')['Client ID'].count().reset_index()
df2.columns= ['Loyalty Classification', 'Num of Clients']
df2.head()
```

```python
df3= df.groupby('Nationality')['Client ID'].count().reset_index()
df3.columns= ['Nationality', 'Num of Clients']
df3.head()
```

```python
df['Joined Bank'] = pd.to_datetime(df['Joined Bank'], dayfirst=True)
```

```python
df.head()
```

```python
df['Year_Joined'] = df['Joined Bank'].dt.year
```

```python
df.head()
```

```python
df4= df.groupby('Year_Joined')['Client ID'].count().reset_index()
df4.columns= ['Year_Joined','Num of Clients']
df4
```

```python
bins = [0, 18, 30, 45, 60, 100]  
labels = ['0-17', '18-29', '30-44', '45-59', '60+']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
```

```python
df.head()
```

```python
df5= df.groupby('Age_Group')['Client ID'].count().reset_index()
df5.columns= ['Age_Group','Num of Clients']
df5
```

```python
df6= df.groupby('Fee Structure')['Estimated Income'].mean().reset_index()
df6.columns= ['Fee Structure', 'Mean Estimated Income']
df6.head()
```

```python
df7= df.groupby('Loyalty Classification')['Estimated Income'].mean().reset_index()
df7.columns= ['Loyalty Classification', 'Mean Estimated Income']
df7.head()
```

```python
df8= df.groupby('Nationality')['Estimated Income'].mean().reset_index()
df8.columns= ['Nationality', 'Mean Estimated Income']
df8.head()
```

```python
df9=df.groupby('GenderId')['Client ID'].count().reset_index()
df9.columns =['Gender','Num of Clients']
df9
```

```python
df10=df.groupby('GenderId')['Estimated Income'].mean().reset_index()
df10.columns =['Gender','Mean Estimated Income']
df10
```

```python
df11=df.groupby('Properties Owned')['Client ID'].count().reset_index()
df11.columns =['Properties Owned','Num of Clients']
df11
```

```python
df12=df.groupby('Properties Owned')['Estimated Income'].mean().reset_index()
df12.columns =['Properties Owned','Mean Estimated Income']
df12.sort_values(by='Mean Estimated Income', ascending=False)
```

```python
df['Loan_to_Deposit_Ratio'] = df['Bank Loans'] / df['Bank Deposits']
df['Credit_Card_Utilization'] = df['Credit Card Balance'] / df['Amount of Credit Cards'].replace(0, pd.NA)
risk_map = {
    1: 'Very Low Risk',
    2: 'Low Risk',
    3: 'Medium Risk',
    4: 'High Risk',
    5: 'Very High Risk'
}
df['Risk_Profile'] = df['Risk Weighting'].map(risk_map)
df.head()
```

```python
df13=df.groupby('Risk_Profile')['Client ID'].count().reset_index()
df13.columns =['Risk_Profile','Num of Clients']
df13
```

```python
df14=df.groupby('Risk_Profile')['Estimated Income'].mean().reset_index()
df14.columns =['Risk_Profile','Mean Estimated Income']
df14.sort_values(by='Mean Estimated Income', ascending= False)
```

# Observations
#### 1. Client Distribution by Fee Structure: The High Fee Structure category attracts the largest number of clients.
#### 2. Loyalty Classification: The Jade Loyalty Classification group has the highest client count.
#### 3. Nationality: The majority of clients are from Europe.
#### 4. Year of Joining: Most clients joined the bank in 2020, indicating a peak acquisition year.
#### 5. Age Groups: Clients aged 60+ years form the largest age segment.
#### 6. Income by Fee Structure: Clients in the High Fee Structure group report the highest average estimated income.
#### 7. Income by Nationality: African clients have the highest mean estimated income compared to other regions.
#### 8. Gender Distribution: Gender 2 clients outnumber Gender 1 clients.
#### 9. Property Ownership: Most clients own two properties.
#### 10. Income by Property Ownership: Surprisingly, clients with no properties report a higher mean estimated income than property owners.
#### 11. Risk & Income: Low-risk clients (Risk Weighting = 1–2) have the highest mean estimated income.

# Conclusions
#### 1. Most clients choose the High Fee Structure and the Jade Loyalty program.
#### 2. The majority of clients are from Europe.
#### 3. Most clients joined the bank in 2020.
#### 4. The 60+ age group has the largest number of clients.
#### 5. High Fee Structure clients earn the highest average income.
#### 6. African clients have the highest mean income.
#### 7. Gender 2 clients are more than Gender 1 clients.
#### 8. Most clients own two properties.
#### 9. Clients with no properties have a higher average income than property owners.
#### 10. Low-risk clients earn the highest average income.
